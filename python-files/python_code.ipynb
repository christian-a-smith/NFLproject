```python
master_df = master_df.drop(columns = 'id')
master_df
```

```python
cleaned_overunders=master_df["overunder"].str.split("(").str[0].astype(float)
cleaned_overunders
```

```python
master_df["overunder"]=cleaned_overunders
master_df
```

```python
#Create code to show the overs that hit for each team, and the unders that hit for each team
```

```python
#Initially start the counter at 0 for each team
overs = {}
for team in nfl_teams:    
    overs[team]=0
unders = {}
for team in nfl_teams:    
    unders[team]=0
```

```python
#iterate through the rows. If the combined point total is greater than the overunder, add 1 to both the home and away team's counter.
#otherwise, if that number is less than the overunder, add 1 to the home and away team's counter for the unders dictionary.
#some results will be pushes (point total exactly equals overunder), and we shouldn't include these.
for index, row in master_df.iterrows():
    if (row.points_home + row.points_away) > row.overunder:
        overs[row.away_team] += 1
        overs[row.home_team] += 1
    elif (row.points_home + row.points_away) < row.overunder:
        unders[row.away_team] += 1
        unders[row.home_team] += 1
```

```python
#Convert these dictionaries into dataframes so we can visualize them
```

```python
overs_df = pd.DataFrame(list(overs.items()), columns = ["Team", "Overs hit"])
overs_df
```

```python
unders_df = pd.DataFrame(list(unders.items()), columns = ["Team", "Unders hit"])
unders_df
```
